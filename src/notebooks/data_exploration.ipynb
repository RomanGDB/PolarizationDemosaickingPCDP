```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Explorarndo el Dataset

```python
dataset_path = '../../dataset'
dataset_subdirs = [name for name in os.listdir(dataset_path)]
print(f'Se cuenta con un total de {len(dataset_subdirs)} imágenes RGB polarizadas a 0°, 45°, 90° y 135° respectivamente.', end = '\n\n')
print(f'Imágenes: {dataset_subdirs}')
```

### Leer una imagen

Las imágenes leídas de esta manera están en formato RGB. Sus píxeles están en el rango [0,255], y son del tipo del numpy.uint8.

```python
img_folder = random.choice(dataset_subdirs)

I_0 = cv2.cvtColor(cv2.imread(f'{dataset_path}/{img_folder}/I0.png'), cv2.COLOR_BGR2RGB)
I_45 = cv2.cvtColor(cv2.imread(f'{dataset_path}/{img_folder}/I45.png'), cv2.COLOR_BGR2RGB)
I_90 = cv2.cvtColor(cv2.imread(f'{dataset_path}/{img_folder}/I90.png'), cv2.COLOR_BGR2RGB)
I_135 = cv2.cvtColor(cv2.imread(f'{dataset_path}/{img_folder}/I135.png'), cv2.COLOR_BGR2RGB)
```

```python
print(f'Imagen: {img_folder}')
print(f'{I_0.shape=}\n{I_45.shape=}\n{I_90.shape=}\n{I_135.shape=}')
```

```python
images = [I_0, I_45, I_90, I_135]
titles = [r'$I_0$', r'$I_{45}$', r'$I_{90}$', r'$I_{135}$']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten() # facilita la iteración
plt.suptitle(f'Imagen: {img_folder}')

for i, ax in enumerate(axes):
    ax.imshow(images[i])
    ax.set_title(f'{titles[i]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
